```python
import numpy as np
import time
# 1. 파이썬 리스트를 사용한 경우
my_list = list(range(1_000_000))
start_time = time.time()
result_list = [x * 2 for x in my_list]
end_time = time.time()
print(f"파이썬 리스트 실행 시간: {end_time - start_time:.5f}초")
```

```python
my_array = np.arange(1_000_000)
start_time = time.time()
result_array = my_array * 2
end_time = time.time()
print(f"넘파이 배열 실행 시간: {end_time - start_time:.5f}초")
```

```python
distances = np.array([1, 2, 3, 5, 8])

windspeeds = np.array([0, 5, 10, 7, 12])
base_usage_rate = 10

additional_usage_rate = windspeeds
total_usage_rate = base_usage_rate + additional_usage_rate
battery_consumption = distances * total_usage_rate
print("비행 거리 (km):", distances)
print("풍속 (m/s):", windspeeds)
print("총 소모율 (%/km):", total_usage_rate)
print("예상 배터리 소모량 (%):", battery_consumption)
```

```python
import pandas as pd
student_data = {
'name': ['Kim', 'Lee', 'Park'],
'test': [85, 92, 78],
'assign1': [88, 95, 80],
'assign2': [91, 89, 82]
}
df = pd.DataFrame(student_data)
display(df)
data_list = [ # Define data_list here
    ['Kim', 85, 88, 91],
    ['Lee', 92, 95, 89],
    ['Park', 78, 80, 82]
]

column_names = ['name', 'test', 'assign1', 'assign2']
df_from_list = pd.DataFrame(data_list, columns=column_names)
display(df_from_list)
```

```python
df['test'] = [85, 92, 78]
for index, row in df.iterrows():
 print(f"인덱스 {index}: {row['name']} 학생의 시험 점수는 {row['test']}점 입니다.")
```

```python
df['sum'] = df['test'] + df['assign1'] + df['assign2']
display(df)
```

```python
high_performers = df[df['sum'] >= 260]
display(high_performers)
```

```python
top_students_summary = df.loc[df['sum'] >= 260, ['name', 'sum']]
display(top_students_summary)
```
